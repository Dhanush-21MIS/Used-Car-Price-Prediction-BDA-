```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv('car_price.csv')
```

```python
# Clean and convert necessary columns
data['kms_driven'] = data['kms_driven'].str.replace(' kms', '').str.replace(',', '').astype(float)
data['engine'] = data['engine'].str.replace(' cc', '').astype(float)
data['Seats'] = data['Seats'].str.replace(' Seats', '').astype(int)
```

```python
# Set style for seaborn
sns.set_style("whitegrid")
```

```python
# Plot 1: Kilometers Driven Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['kms_driven'], bins=30, kde=True)
plt.title('Kilometers Driven Distribution')
plt.xlabel('Kilometers Driven')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot 2: Fuel Type Distribution
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='fuel_type', palette='Set2')
plt.title('Fuel Type Distribution')
plt.xlabel('Fuel Type')
plt.ylabel('Count')
plt.show()
```

```python
# Plot 3: Transmission Type Distribution
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='transmission', palette='Set3')
plt.title('Transmission Type Distribution')
plt.xlabel('Transmission')
plt.ylabel('Count')
plt.show()
```

```python
# Plot 4: Ownership Status Distribution
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='ownership', palette='Set1')
plt.title('Ownership Status Distribution')
plt.xlabel('Ownership Status')
plt.ylabel('Count')
plt.show()
```

```python
# Plot 5: Year of Manufacture Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['manufacture'], bins=20, kde=True)
plt.title('Year of Manufacture Distribution')
plt.xlabel('Year of Manufacture')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot 6: Engine Capacity Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['engine'], bins=30, kde=True)
plt.title('Engine Capacity (cc) Distribution')
plt.xlabel('Engine Capacity (cc)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot 7: Seats Distribution
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Seats', palette='Set1')
plt.title('Seats Distribution')
plt.xlabel('Seats')
plt.ylabel('Count')
plt.show()
```

```python
# 1. Box Plot of Kilometers Driven by Fuel Type
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='fuel_type', y='kms_driven')
plt.title('Kilometers Driven by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Kilometers Driven')
plt.show()
```

```python
# 4. Scatter Plot of Car Price vs Engine Capacity
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='engine', y='car_prices_in_rupee', hue='fuel_type')
plt.title('Car Price vs Engine Capacity')
plt.xlabel('Engine Capacity (cc)')
plt.ylabel('Car Price (Rupees)')
plt.show()
```

```python
# 5. Box Plot of Car Price by Ownership Status
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='ownership', y='car_prices_in_rupee', palette="Set1")
plt.title('Car Price by Ownership Status')
plt.xlabel('Ownership')
plt.ylabel('Car Price (Rupees)')
plt.show()
```

```python
# 6. Pair Plot of Numerical Variables
sns.pairplot(data[['car_prices_in_rupee', 'kms_driven', 'engine', 'manufacture']])
plt.suptitle('Pair Plot of Numerical Variables', y=1.02)
plt.show()
```
